```python
import pandas as pd
import numpy as np
df = pd.read_csv("bowlersLabel.csv")
df.head()
```

```python
ndf = df.drop(['Player Name','Span','Mat','Inns','Overs','Mdns','Runs','Wkts','BBI','4WH','5WH','Team name','Nation'],axis=1)
ndf
```

```python
from sklearn.preprocessing import LabelEncoder
le_status = LabelEncoder()
ndf['status_n'] = le_status.fit_transform(ndf['status'])
ndf
```

# expensive_bowler => 0, good_bowler=> 1, key_bowler => 2

```python
ndf2 = ndf.drop("status",axis=1)
ndf2.head()
```

```python
X = ndf2[['Ave','Econ','SR']]
y = ndf2[['status_n']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
X_test
```

```python
print("Percentage of model's accuracy is: ",model.score(X_test,y_test)*100,"%")
```

# expensive_bowler => 0, good_bowler=> 1, key_bowler => 2

```python
import tkinter as tk
from tkinter import *

screen = Tk()

def pred():
    ave = average.get()
    ec = economy.get()
    st = strikeRate.get()
    cat = model.predict([[ave,ec,st]])
    if cat == 0:
        catt = "expensive_bowler"
        output.config(text="Category is: "+catt,font=("calibri",12))
    elif cat == 1:
        catt = "good_bowler"
        output.config(text="Category is: "+catt,font=("calibri",12))
    else:
        catt = "key_bowler"
        output.config(text="Category is: "+catt,font=("calibri",12))
        
def delt():
    if average.get()=="" and economy.get() == "" and strikeRate.get() == "":
        msg.showerror("Error","Empty Field")
    else:
        average.set("")
        economy.set("")
        strikeRate.set("")
    
screen.geometry("600x600")
screen.title("Bowler's Class")
header = Label(text="FIND BOWLER'S CLASS BY FILLING REQUIRED FIELDS",font=("calibri",20),bg="#ffad00",fg="black",width="100",
              height="3")
header.pack()

ave = Label(text="Average: ",font=("calibri",15),bg="black",fg="#ffad00")
ave.place(x=10,y=150)
eco = Label(text="Economy: ",font=("calibri",15),bg="black",fg="#ffad00")
eco.place(x=10,y=250)
sr = Label(text="Strike Rate: ",font=("calibri",15),bg="black",fg="#ffad00")
sr.place(x=10,y=350)

average = DoubleVar()
economy = DoubleVar()
strikeRate = DoubleVar()

ave_entry = Entry(textvariable = average, width="70",bg="#ffad00",fg="black")
eco_entry = Entry(textvariable = economy, width="70",bg="#ffad00",fg="black")
sr_entry = Entry(textvariable = strikeRate, width="70",bg="#ffad00",fg="black")

ave_entry.place(x=10,y=190)
eco_entry.place(x=10,y=290)
sr_entry.place(x=10,y=390)

find = Button(text="Click to Predict for Category", width="25",height=2,bg="black",fg="#ffad00",font=("calibri",12),
             command = pred)
find.place(x = 45, y = 450)
delete = Button(text="Clear all fields", width="25",height=2,bg="black",fg="#ffad00",font=("calibri",12),
             command = delt)
delete.place(x = 290, y = 450)
output = Label(fg="black",bg="#ffad00",width="100",height=3)
output.place(x=0,y=550)
screen.mainloop()
```
