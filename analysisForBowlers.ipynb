# Key bowlers from Pakistan

```python
keyBowlersPak = df[(df['status'] == 'key_bowler') & (df['Nation'] == 'Pakistan')]
keyBowlersPak
```

```python
#defining plotting region (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.2, wspace=1)
sns.set(rc = {'figure.figsize':(15,8)})
#creating barplot in each subplot
sns.barplot(data=keyBowlersPak, x='Inns', y='Player Name', ax=axes[0,0])
sns.barplot(data=keyBowlersPak, x='Overs', y='Player Name', ax=axes[0,1])
sns.barplot(data=keyBowlersPak, x='Mdns', y='Player Name', ax=axes[1,0])
sns.barplot(data=keyBowlersPak, x='Runs', y='Player Name', ax=axes[1,1])
```

```python
#defining plotting region (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.5, wspace=1)
#creating barplot in each subplot
sns.barplot(data=keyBowlersPak, x='Wkts', y='Player Name', ax=axes[0,0])
sns.barplot(data=keyBowlersPak, x='Ave', y='Player Name', ax=axes[0,1])
sns.barplot(data=keyBowlersPak, x='Econ', y='Player Name', ax=axes[1,0])
sns.barplot(data=keyBowlersPak, x='SR', y='Player Name', ax=axes[1,1])
```

```python
#defining plotting region (1 row, 2 columns)
fig, axes = plt.subplots(1, 2)
fig.subplots_adjust(hspace=0.5, wspace=1)
#creating barplot in each subplot
sns.barplot(data=keyBowlersPak, x='4WH', y='Player Name', ax=axes[0])
sns.barplot(data=keyBowlersPak, x='5WH', y='Player Name', ax=axes[1])
```

```python
anno1 = keyBowlersPak['Player Name']
plt.xlabel("Average",fontsize=15)
plt.ylabel("Strike Rate",fontsize=15)
av = keyBowlersPak['Ave']
sr = keyBowlersPak['SR']
for i,type in enumerate(anno1):
    x = av.iloc[i]
    y = sr.iloc[i]
    plt.scatter(x, y, marker='.', color='red')
    plt.text(x, y, type, fontsize=15)
plt.figure(figsize=(6,6))    
plt.show()
```

```python
annoo = keyBowlersPak['Player Name']
plt.xlabel("Innings",fontsize=15)
plt.ylabel("Average",fontsize=15)
avv = keyBowlersPak['Inns']
srr = keyBowlersPak['Ave']
for i,type in enumerate(annoo):
    x = avv.iloc[i]
    y = srr.iloc[i]
    plt.scatter(x, y, marker='.', color='red')
    plt.text(x, y, type, fontsize=15)
plt.figure(figsize=(6,6))    
plt.show()
```

```python
annooo = keyBowlersPak['Player Name']
plt.xlabel("Innings",fontsize=15)
plt.ylabel("Strike Rate",fontsize=15)
avvv = keyBowlersPak['Inns']
srrr = keyBowlersPak['SR']
for i,type in enumerate(annooo):
    x = avvv.iloc[i]
    y = srrr.iloc[i]
    plt.scatter(x, y, marker='.', color='red')
    plt.text(x, y, type, fontsize=15)
plt.figure(figsize=(6,6))    
plt.show()
```
