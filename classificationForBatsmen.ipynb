```python
import pandas as pd
import numpy as np
df = pd.read_csv("batsmenLabel.csv")
df.head()
```

```python
ndf = df.drop(['Player Name','Span','Mat','NO','Runs','Highest Score','Balls Faced','100s','50s','0s','4s','6s',
              'Team name','Nation'],axis=1)
ndf
```

```python
from sklearn.preprocessing import LabelEncoder
le_status = LabelEncoder()
ndf['status_n'] = le_status.fit_transform(ndf['status'])
ndf
```

# power_hitter => 2 , good_performer => 1,  average_performer => 0

```python
ndf2 = ndf.drop("status",axis=1)
ndf2.head()
```

```python
X = ndf2[['Inns','Ave','Strike Rate']]
y = ndf2[['status_n']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
X_test
```

```python
print("Percentage of model's accuracy is: ",model.score(X_test,y_test)*100,"%")
```

# power_hitter => 2 , good_performer => 1, average_performer => 0

```python
import tkinter as tk
from tkinter import *

screen = Tk()

def pred():
    inns = innings.get()
    av = average.get()
    st = strikeRate.get()
    cat = model.predict([[inns,av,st]])
    if cat == 0:
        catt = "average_performer"
        output.config(text="Category is: "+catt,font=("calibri",12))
    elif cat == 1:
        catt = "good_performer"
        output.config(text="Category is: "+catt,font=("calibri",12))
    else:
        catt = "power_hitter"
        output.config(text="Category is: "+catt,font=("calibri",12))
        
def delt():
    if innings.get()=="" and average.get() == "" and strikeRate.get() == "":
        msg.showerror("Error","Empty Field")
    else:
        innings.set("")
        average.set("")
        strikeRate.set("")        
screen.geometry("600x700")
screen.title("Player's Class")
header = Label(text="FIND PLAYER'S CLASS BY FILLING REQUIRED FIELDS",font=("calibri",20),bg="#ffad00",fg="black",width="100",
              height="3")
header.pack()

inn = Label(text="Number of innings: ",font=("calibri",15),bg="black",fg="#ffad00")
inn.place(x=10,y=150)
ave = Label(text="Average: ",font=("calibri",15),bg="black",fg="#ffad00")
ave.place(x=10,y=250)
sr = Label(text="Strike Rate: ",font=("calibri",15),bg="black",fg="#ffad00")
sr.place(x=10,y=350)

innings = IntVar()
average = DoubleVar()
strikeRate = DoubleVar()

inn_entry = Entry(textvariable = innings, width="70",bg="#ffad00",fg="black")
ave_entry = Entry(textvariable = average, width="70",bg="#ffad00",fg="black")
sr_entry = Entry(textvariable = strikeRate, width="70",bg="#ffad00",fg="black")

inn_entry.place(x=10,y=190)
ave_entry.place(x=10,y=290)
sr_entry.place(x=10,y=390)

find = Button(text="Click to Predict for Category", width="25",height=2,bg="black",fg="#ffad00",font=("calibri",12),
             command = pred)
find.place(x = 90, y = 450)
delete = Button(text="Clear all fields", width="25",height=2,bg="black",fg="#ffad00",font=("calibri",12),
             command = delt)
delete.place(x = 290, y = 450)
output = Label(fg="black",bg="#ffad00",width="100",height=3)
output.place(x=0,y=550)
screen.mainloop()
```
