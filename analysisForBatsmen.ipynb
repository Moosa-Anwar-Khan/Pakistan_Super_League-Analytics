# Power hitters from Pakistan

```python
powerHittersPak = df[(df['status'] == 'power_hitter') & (df['Nation'] == 'Pakistan')]
powerHittersPak
```

```python
#defining plotting region (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.5, wspace=0.5)
sns.set(rc = {'figure.figsize':(15,8)})
#creating barplot in each subplot
a = sns.barplot(data=powerHittersPak, x='Inns', y='Player Name', ax=axes[0,0])
a.set(xlabel="Innings",ylabel="Player Name")
b = sns.barplot(data=powerHittersPak, x='NO', y='Player Name', ax=axes[0,1])
b.set(xlabel="NO. of times being notout",ylabel="Player Name")
c = sns.barplot(data=powerHittersPak, x='Runs', y='Player Name', ax=axes[1,0])
c.set(xlabel="Runs",ylabel="Player Name")
d = sns.barplot(data=powerHittersPak, x='Ave', y='Player Name', ax=axes[1,1])
d.set(xlabel="Average",ylabel="Player Name")
```

```python
#defining plotting region (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.5, wspace=1)
#creating barplot in each subplot
bg1 = sns.barplot(data=powerHittersPak, x='Balls Faced', y='Player Name', ax=axes[0,0])
bg2 = sns.barplot(data=powerHittersPak, x='Strike Rate', y='Player Name', ax=axes[0,1])
bg3 = sns.barplot(data=powerHittersPak, x='100s', y='Player Name', ax=axes[1,0])
bg4 = sns.barplot(data=powerHittersPak, x='50s', y='Player Name', ax=axes[1,1])
```

```python
#defining plotting region (1 row, 2 columns)
fig, axes = plt.subplots(1, 2)
fig.subplots_adjust(hspace=0.5, wspace=1)
#creating barplot in each subplot
sns.barplot(data=powerHittersPak, x='4s', y='Player Name', ax=axes[0])
sns.barplot(data=powerHittersPak, x='6s', y='Player Name', ax=axes[1])
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data=powerHittersPak, x='0s', y='Player Name')
plt.xlabel("Ducks",fontsize=15)
plt.ylabel("Player Name",fontsize=15)
```

#As i want to select those power hitters in my team who have good strike rate and good average so we will draw scatter plot 
#with average on x-axis and strike rate on y-axis. we will select 3 power hitter for our team

```python
anno1 = powerHittersPak['Player Name']
plt.xlabel("Average",fontsize=15)
plt.ylabel("Strike Rate",fontsize=15)
av = powerHittersPak['Ave']
sr = powerHittersPak['Strike Rate']
for i,type in enumerate(anno1):
    x = av.iloc[i]
    y = sr.iloc[i]
    plt.scatter(x, y, marker='.', color='red')
    plt.text(x, y, type, fontsize=15)
plt.figure(figsize=(6,6))    
plt.show()
```

```python
powerHittersSelected = powerHittersPak[(powerHittersPak['Player Name'] == 'Asif Ali') | 
                                      (powerHittersPak['Player Name'] == 'Sharjeel Khan') | 
                                      (powerHittersPak['Player Name'] == 'Haider Ali')]
powerHittersSelected
```

Making CSV file of power hitter selected for team

```python
powerHittersSelected.to_csv("powerHittersSelected.csv",index=False)
```
